```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

from plotly import express as exp

from plotly import graph_objects as go

import datetime as dt

from joblib import dump, load
```

```python
lar_df = pd.read_csv(r"D:\py_projects\san_francisco_public_data\san_francisco_crimes_data_from_2018\larceny_cases_from_2018_dropped.csv")


neigh_df = pd.read_csv("./san_francisco_crimes_data_from_2018/address_list_neigh.csv")


lar_df.shape, neigh_df.shape

```

```python
lats = lar_df.latitude.values.tolist()

longs = lar_df.longitude.values.tolist()
```

```python
zips = []

neighs = []

for i in range(len(lats)):
    
    try:
        
        zips.append(neigh_df[(neigh_df.latitude_ == lats[i]) & (neigh_df.longitude_ == longs[i])]["zipcode"].values[0])
        
    
    except:
        
        zips.append(np.nan)
    
    
    try:
        neighs.append(neigh_df[(neigh_df.latitude_ == lats[i]) & (neigh_df.longitude_ == longs[i])]["neighborhood"].values[0])
        
        
    except:
        
        neighs.append(np.nan)
```

```python
len(zips)
```

```python
lar_df["zipcode"] = zips

lar_df["neighborhood"] = neighs
```
